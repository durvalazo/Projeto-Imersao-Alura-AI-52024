```python
# Instalar Google Generative AI
!pip install -q -U google-generativeai

import google.generativeai as genai
from google.colab import userdata
api_key = userdata.get('SECRET_KEY')
# Google API key
GOOGLE_API_KEY = "AIzaSyAhgeVeFp3Q5ii6DZCc5sSvqKQdzrIncSQ"
genai.configure(api_key=api_key)

# Generation Ajustes
generation_config = {
  "candidate_count": 1,  # Numero de Respostas
  "temperature": 0.7,  # Creatividade 0-1
}

# Safety ajustes
safety_settings = {
  "HARASSMENT": "BLOCK_NONE",
  "HATE": "BLOCK_NONE",
  "DANGEROUS": "BLOCK_NONE",
}

#  modelo
model = genai.GenerativeModel(
  model_name="gemini-1.0-pro",
  generation_config=generation_config,
  safety_settings=safety_settings,
)

# voz (falta integracao externa)
def falar(texto):
  # Text-to-Speech)
  print("Voz:", texto)  # simulacao temporaria

# Go
chat = model.start_chat(history=[])

print("\033[1mOlá! Eu sou seu assistente virtual. Para melhor te atender, preciso saber sua idade. Qual é a sua idade?\033[0m")

while True:
  # maiusculas ajudar Idoso
  prompt = input("Você: ").upper()

  # Fim chat
  if prompt == "FIM":
    break

  try:
    # idade
    age = int(prompt)
    if age < 0:
      raise ValueError("A idade não pode ser negativa.")
  except ValueError:
    print("\033[1mIdade inválida. Por favor, digite um número inteiro positivo.\033[0m")
    continue

  # Tamnho de letra Basado em idade
  if age > 80:
    large_font = True
    print("\033[1mÓtimo! Para facilitar a leitura, usarei um tamanho de letra maior.\033[0m")
  else:
    large_font = False

  # Primeira vez o Usuario
  first_time = input("\033[1mÉ a sua primeira vez usando o Chatbot Gemini? (SIM/NÃO) ").upper()

  if first_time == "SIM":
    # Bemvindo Novo usuario
    print("\033[1mBem-vindo ao Chatbot Gemini! Estou aqui para te ajudar em diversas tarefas.\033[0m")

    # Dicas
    print("\033[1mAlgumas dicas para começar:\033[0m")
    print("1. Faça perguntas abertas e diretas.")
    print("2. Use linguagem natural como se estivesse conversando com um amigo.")
    print("3. Explore diferentes comandos e funcionalidades para descobrir o que posso fazer.")

  # Ajuda
  print("\033[1mEm que posso te ajudar hoje?\033[0m")

  # Ajuda ate ele dizer  "FIM"
  while True:
    # Maiusculas ver melhor
    user_query = input("Você: ").upper()

    #  "FIM"
    if user_query == "FIM":
      break

    # Pergunta y resposta Gemini
    response = chat.send_message(user_query)

    # Voz
    # falar(response.text)

    # Resposta com Letra Maior
    if large_font:
      print(f"\033[1mAssistente: {response.text.upper()}\033[0m")
    else:
      print(f"Assistente: {response.text.upper()}")
```
